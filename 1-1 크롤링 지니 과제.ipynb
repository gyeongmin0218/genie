```python
import time
import requests
from bs4 import BeautifulSoup


header = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)'}
info =[];
for i in range(4):
  request = requests.get(f'https://genie.co.kr/chart/top200?ditc=D&ymd=20240520&hh=14&rtm=Y&pg={i+1}', headers=header)
  soup= BeautifulSoup(request.text)



  for i in range(50):
    infos = soup.findAll('td', {'class' : 'info'})
    info.append(infos[i].text.strip().split('\n'))

  time.sleep(1);


for i in range(len(info)):
    print(f'{i+1}위 {info[i][0]} - {info[i][1]}')


```
